# Distribution Plots

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#Dist Plot - Allows us to show distribution of uni-variate set of observations : univariate refers to single variable
#Dist plot refers to distribution plots
```

```python
sns.distplot(tips['total_bill'])
#in dist plot we pass single column of database
```

```python
#The curve over distpot is called kde, we can set it false to take it off
sns.distplot(tips['total_bill'], kde= False)
```

```python
#Change number of bins
sns.distplot(tips['total_bill'], kde= False, bins=30)
```

```python
#If we increase bin size to 100, it will show more minute details
sns.distplot(tips['total_bill'], kde= False, bins=100)
```

# Joint Plot

```python
#It allows us to match up two distplots by varied plots, i.e., we can combine two different distribution plots
```

```python
sns.jointplot(x = 'total_bill', y= 'tip', data = tips)
#We cannot use data parametre like (data = tips )in distplot as it will generate errors.
#Joint plot takes two variables and data parametre.
```

```python
#The above joint plot has two distribution plots and one scatter plot.
```

```python
#We can change style of scatter plot using kind argument.
sns.jointplot(x = 'total_bill', y= 'tip', data = tips, kind = 'hex')
```

```python
#Another argument we can use is reg, which stands for regression
sns.jointplot(x = 'total_bill', y= 'tip', data = tips, kind = 'reg')
```

```python
#We can set scatter plot to 2D KDE, it shows the density where the data points meet up the most.
sns.jointplot(x = 'total_bill', y= 'tip', data = tips, kind = 'kde')
```

```python
#We are mainly going to use joint plot with default scatter bcz it is easiest to read and gives us quite a bit of information.
sns.jointplot(x = 'total_bill', y= 'tip', data = tips)
```

# Pair Plot

It is going to establish pair wise relationship across an entire data frame, atleast for the numerical columns. 
It will also support a clor hue argument for categorical clumns

```python
sns.pairplot(tips)
```

For total_bill vs total_bill, tip vs tip, and size vs size i.e, variable vs same_variable, we get result as histogram 
instead of scatter plot

```python
#Add hue argument, we will pass categorical column i.e, which is not numrical or continous but is actually a category.
#eg,sex column which contains male and female categories
sns.pairplot(tips, hue='sex')
```

In above graphs, data points are colored based on sex category.

```python
#We can pass color palette argument
sns.pairplot(tips, hue = 'sex', palette='coolwarm')
# Search for more palette on seaborn website
```

# Rug Plot

```python
#Like dist plot we will pass single column here.
```

```python
sns.rugplot(tips['total_bill'])
```

```python
#The rugplot is not working above, as there is change in code with update, below is the new code:
#kde plot plots univariate or bivariate kernal density estimate
sns.kdeplot(tips['total_bill'], ls="")
sns.rugplot(tips['total_bill'])
```

```python
#It just draws a dash mark for every point of every univariate variable 
```

Difference between rugplot and histogram

```python
sns.distplot(tips['total_bill'], kde = False)
```

Histogram essentially has bins and it counts how many dashes or data points are in that bin and shows it as a number on y axis.
More dashes in rugplot at a particular place means the bin is gonna be high and less dashes mean bin is low at that 
particular place.

```python
#to build kde line using rugplot, we use kde plot
```

```python
#Let us build a basic kde plot
sns.kdeplot(tips['total_bill'])
```

```python
#Let us build kde plot with rug plot.
sns.kdeplot(tips['total_bill'], color='r')
sns.rugplot(tips['total_bill'])
```
