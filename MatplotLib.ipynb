## Matplot Library

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
```

```python
# Function
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Line Graph')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'y')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'r')
axes1.set_title('Larger Plot')

axes2.plot(y,x, 'b')
axes2.set_title('Smaller Plot')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,1,1])
axes2 = fig.add_axes([0.2,0.6,0.5,0.4])
axes1.plot(x,y)
axes2.plot(y,x)
```

```python
#instead of creating figure and axis, we can do same with subplots in an easy way.
fig, axes = plt.subplots(1,2)
#plt.subplot automatically creates figures and does fig.add_axes([0.1,0.1,1,1]) itself with no need to supply axis lbwh
```

```python
fig, axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
#Create 3 rows and 3 cols sub-plots
fig, axes = plt.subplots(3,3)
#To get rid of overlapping of sub-plots, use:
plt.tight_layout()
```

```python
#Plot on 2 dimensional sub-plots.
fig, axes = plt.subplots(3,3)
plt.tight_layout()
#Plot on 2 dimensional sub-plots.
axes[0,1].plot(x,y)
```

```python
#Set title in sub-plot axis.
fig, axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[0].set_title("First Plot")
axes[1].plot(y,x)
axes[1].set_title("Second Plot")
```

# Figure size, Aspect ratio and Dpi

```python
#figsize sets figure size with parametres width and hieght
fig = plt.figure(figsize = (4,4))
ax = fig.add_axes([0.1,0.1,1,1])
ax.plot(x,y)
```

```python
#ame for subplots, figsize sets figure size with parametres width and hieght
fig, axes = plt.subplots(figsize = (4,4))
#We will not use indexing as we have only one axis, and indexing will give error.
axes.plot(x,y,'g')
```

```python
#Multiple sub-plots
fig, axes = plt.subplots(2,1,figsize = (8,4))
axes[0].plot(x,y)
axes[0].set_title('First plot')
axes[0].set_xlabel('X Label')
axes[0].set_ylabel('Y Label')

axes[1].plot(x,y)
axes[1].set_title('Second plot')
axes[1].set_xlabel('X Label')
axes[1].set_ylabel('Y Label')

plt.tight_layout()
```

```python
#Save figure, you can provide any extension, and set dpi
fig.savefig('my_picture.png', dpi = 100)
#setting dpi to 200 will double its size.
```

# Legends

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label = 'X Squared')
ax.plot(x,x**3, label = 'Y cubed')

ax.legend()
```

```python
#Set legend location
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label = 'X Squared')
ax.plot(x,x**3, label = 'Y cubed')
#We can specify locations like best, upper right,lower right, center, upper center
ax.legend(loc = 'upper right')
```

```python
#Specify location of legend using left and bottom
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label = 'X Squared')
ax.plot(x,x**3, label = 'Y cubed')

ax.legend(loc =(0.1,0.1))
```

# Setting colors with matplotlib

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color = 'orange')
```

```python
# Setting color code using RGB hex code
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color = '#FF0000')
```

```python
#Shortcut for color code
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,'#000000')
# (or)
# ax.plot(x,y,'r') where r stands for red.
```

# Line width and line style

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
#Default line width is 1
ax.plot(x,y, color = '#FF0000', linewidth = 2)
#line width 2 will double its size
#shortcut for linewidth is lw
```

```python
#Set transparency of line, using alpha attribute
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = '#FF0000', lw = 5, alpha = 0.5)
#alpha = 0.5 will make it more transparent
```

# Line Style

```python
# Make line of --
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = '#FF0000', lw = 5, linestyle = '--')
```

```python
# Make line of -.
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = '#FF0000', lw = 5, linestyle = '-.')
```

```python
# Make line of Steps
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
#ax.plot(x,y, color = '#FF0000', lw = 5, linestyle = 'steps') is deprecated and new approach is given below. 
line, = ax.plot(x,y, color = '#FF0000', lw = 5)
line.set_drawstyle("steps")
#line._invalidx = True
plt.show()
```

```python
#Shortcut for line style
# Make line of --
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = '#FF0000', lw = 5, ls = '-')
```

# Markers

```python
len(x)
```

```python
#Let us mark these 11 points and check where they lie on this plot
#Shortcut for line style
# Make line of --
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
#Check matplot lib for different markers
#Capital O will give error, make sure to use lower case o
ax.plot(x,y, color = '#FF0000', lw = 2, ls = '-', marker = 'o')
#It marks each point on x and y
```

```python
#Set marker size with shortcut
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
#Set using attrib markersize = 10, or shortcut ms = 10
ax.plot(x,y, color = '#FF0000', lw = 2, ls = '-', marker = '*', ms = 10)
```

```python
#Set marker color
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = '#FF0000', lw = 2, ls = '-', marker = '*', ms = 20, markerfacecolor = 'green')
```

```python
#Shortcut for markerfacecolor
#Set marker color
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
#Capital O will give error, make sure to use lower case o
ax.plot(x,y, color = '#FF0000', lw = 2, ls = '-', marker = 'o', ms = 20, mfc = 'purple')
```

```python
# Set marker edge/border width
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = '#FF0000', lw = 2, ls = '-', marker = 'o', ms = 20, markerfacecolor = 'green', markeredgewidth = 4)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = '#FF0000', lw = 2, ls = '-', marker = 'o', ms = 20, markerfacecolor = 'green', markeredgewidth = 3, 
        markeredgecolor = 'black')
```

```python
#Shortcut for above code
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = '#FF0000', lw = 2, ls = '-', marker = 'o', ms = 20, mfc = 'green', mew = 3, 
        mec = 'black')
```

```python
#Control Over axis apperance / Zoom to a specific area of plot
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = '#FF0000', lw = 2, ls = '--')
#Lets us say, we want to plot between 0 to 1 on x axis.
ax.set_xlim([0,1])
#Lets us say, we want to plot between 0 to 2 on x axis.
ax.set_ylim([0,2])
```
